```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('lagged.csv')
```

```python
df
```

```python
df=df.dropna()
```

```python
df=df.drop(['datetime_utc'],axis=1)
```

```python
df=df.drop(['Unnamed: 0',' _dewptm',' _fog',' _hail',' _hum',' _pressurem',' _snow',' _rain',' _thunder',' _tornado',' _wdird'],axis=1)
df.columns
```

```python
df.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(df.drop([' _tempm'],axis=1),df[' _tempm'],test_size=0.20)
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
regression_model = LinearRegression()
```

```python
regression_model.fit(X_train, y_train)
y_predicted = regression_model.predict(X_train)
y_predicted
```

```python
rmse = mean_squared_error(y_train, y_predicted)
r2 = r2_score(y_train, y_predicted)
print('Slope:' ,regression_model.coef_)
print('Intercept:', regression_model.intercept_)
print('Root mean squared error: ', rmse)
print('R2 score: ', r2)
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
mean_absolute_error(y_train,y_predicted)
```

```python
plt.scatter(y_predicted,y_train)
```

```python
from sklearn.feature_selection import f_regression
```

```python
print(f_regression(y_predicted.reshape(-1,1),y_train))
```

# VALIDATION

```python
validation_model=LinearRegression()
validation_model.fit(X_test,y_test)
y_tested=validation_model.predict(X_test)
y_tested
rmse = mean_squared_error(y_test, y_tested)
r2 = r2_score(y_test, y_tested)
print('Slope:' ,validation_model.coef_)
print('Intercept:', validation_model.intercept_)
print('Root mean squared error: ', rmse)
print('R2 score: ', r2)
```

```python
mean_absolute_error(y_test,y_tested)
```

```python
plt.scatter(y_tested,y_test,c='m')
```

```python
accuracy = validation_model.score(X_test,y_test)
print("ACCURACY IS:-> "+str(accuracy*100)+' %')
```

```python
from sklearn.feature_selection import f_regression
print(f_regression(y_tested.reshape(-1,1),y_test))
```
